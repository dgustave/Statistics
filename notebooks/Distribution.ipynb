```python
commute_df.set_index(commute_df['State'], inplace= True)
commute_df_cols_to_check= commute_df.columns
commute_df[commute_df_cols_to_check]= commute_df[commute_df_cols_to_check].replace({'minutes':''}, regex=True)
commute_df[commute_df.columns[2]] = commute_df[commute_df.columns[2]].map(lambda x: x.replace('+/-',''))
commute = commute_df.apply(pd.to_numeric, errors='coerce')
commute.drop(columns = 'State', inplace=True)
commute.reset_index(inplace= True)
commute
```
